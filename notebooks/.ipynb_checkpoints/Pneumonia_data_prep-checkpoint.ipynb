```python
## necessary imports
import pandas as pd
import matplotlib.pyplot as plt
import os
import missingno as msno
from pathlib import Path
```

```python
images = pd.read_csv("../output/image_data.csv")
pneumonia = pd.read_csv("../output/pneumonia_only.csv")
```

```python
## converting study_id and subject_id from integer to string

pneumonia[['subject_id','study_id']]=pneumonia[['subject_id','study_id']].astype(str)
```

### Matching Patients in Pneumonia DF to corresponding ChestXray Image

```python
pneumonia['subject_id'] = 'p' + pneumonia['subject_id'].astype(str)
```

```python
pneumonia['study_id'] = 's' + pneumonia['study_id'].astype(str)
```

```python
common = pd.merge(images, pneumonia, on=['study_id'], how='inner')
```

```python
common.head(6)
```

```python
pneumonia_dataset = common[['patient_id','study_id','image_id','image_path','Pneumonia']]
```
